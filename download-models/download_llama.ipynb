```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained("IlyaGusev/saiga_llama3_8b", trust_remote_code=True)
tokenizer.padding_side = 'right'
tokenizer.pad_token = tokenizer.eos_token
tokenizer.add_eos_token = True
tokenizer.add_bos_token = True
```

```python
base_model = AutoModelForCausalLM.from_pretrained(
    "IlyaGusev/saiga_llama3_8b",
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True
)
```

```python
# this one should pass if you downloaded the weights correctly
model = PeftModel.from_pretrained(base_model, "../artifacts/peft_llama")
```

```python
tokenizer.save_pretrained("../artifacts/saiga_llama3_tokenizer")
base_model.save_pretrained("../artifacts/saiga_llama3_base_model")
```
